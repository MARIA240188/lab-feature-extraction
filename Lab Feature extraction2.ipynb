```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_cliente = pd.read_csv(r'C:\Users\Win10\Downloads\we_fn_use_c_marketing_customer_value_analysis.csv')
```

```python
df_cliente = df_cliente.drop(columns=['Customer'])
```

```python
# Seleccionar variables numéricas y categóricas
df_numericas = df_cliente.select_dtypes(include=['int64', 'float64'])
df_categoricas = df_cliente.select_dtypes(include=['object'])
```

1. Open the categoricals variable we created before.

```python
df_categoricas.head()
```

2. Plot all the categorical variables with the proper plot. What can you see?

```python
# Graficar cada variable categórica
for columna in df_categoricas.columns:
    plt.figure(figsize=(10, 6))
    df_categoricas[columna].value_counts().plot(kind='bar')
    plt.title(f'Distribución de la variable categórica {columna}')
    plt.xlabel(columna)
    plt.ylabel('Frecuencia')
    plt.show()
```

3. There might be some columns that seem to be redundant, check their values to be sure. What should we do with them?

```python
# Revisar valores únicos en cada columna categórica
for columna in df_categoricas.columns:
    print(f'Valores únicos en la columna {columna}:\n{df_categoricas[columna].value_counts()}\n')
```

4. Plot time variable. Can you extract something from it?

```python
# Asegurarnos de que la columna de tiempo esté en formato datetime
df_cliente['Effective To Date'] = pd.to_datetime(df_cliente['Effective To Date'])

# Graficar la variable de tiempo
plt.figure(figsize=(10, 6))
df_cliente['Effective To Date'].value_counts().sort_index().plot()
plt.title('Distribución de la variable de tiempo')
plt.xlabel('Fecha')
plt.ylabel('Frecuencia')
plt.show()
```
